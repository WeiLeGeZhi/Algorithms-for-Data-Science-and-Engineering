```python
import tifffile as tf
import numpy as np
from PIL import Image
import math
import time
from sympy import *
import random
from numpy import *
```

```python
def eig_by_hand(M,max_iteration = 100,epsilon = 1e-3):
    A = M
    S = np.eye(len(M))
    for I in range (max_iteration):
        Q,R = np.linalg.qr(A)
        S = S@Q
        A_1 = R@Q
        if (np.linalg.norm(A_1-A,ord = np.inf)<=epsilon):
            A = A_1
            break
        A = A_1
    ev = []
    for i in range(len(M)):
        ev.append(A[i][i])
    eigenvalue = np.array(ev)
    return eigenvalue,S
```

```python
def pic_cpr(route,k):
    start = time.perf_counter()
    img0 = tf.imread(route)
    print("route")
    print(route)
    print("size")
    print(img0.shape)
    #把三维的图片转化为二维的矩阵
    Red = np.zeros((img0.shape[0],img0.shape[1]))
    Green = np.zeros((img0.shape[0],img0.shape[1]))
    Blue = np.zeros((img0.shape[0],img0.shape[1]))
    for i in range (img0.shape[0]):
        for j in range (img0.shape[1]):
            Red[i][j]=img0[i][j][0]
            Green[i][j]=img0[i][j][1]
            Blue[i][j]=img0[i][j][2]
    img1 = np.hstack((Red,Green,Blue))
    #中心化、标准化
    means = np.mean(img1, axis=0).reshape(1, -1)
    stvr = np.std(img1, axis=0).reshape(1, -1)
    img2 = (img1 - means) / stvr
    #求出协方差矩阵并对其进行主成分分析
    S = img2.T@img2
    W, Q = eig_by_hand(S)
    EV_sorted = np.flip(np.argsort(W))
    Q = Q[:, EV_sorted]
    W = W[EV_sorted]
    #根据主成分个数k保存前k个特征向量，并与原矩阵作乘积
    C = img2@Q
    C_to_save = C[:, :k]
    Q_to_save = Q.T[:k, :]
    #根据主成分个数计算压缩率并对图片进行重构
    compact_rate = (C_to_save.shape[0]*C_to_save.shape[1]+Q_to_save.shape[0]*Q_to_save.shape[1])/(img0.shape[0]*img0.shape[1]*img0.shape[2])
    img3 = C_to_save@Q_to_save
    img3 = img3 * stvr + means
    #把的矩阵转化为三维的图片
    img4=np.zeros((img0.shape[0],img0.shape[1],3),dtype = int)
    for i in range (img0.shape[0]):
        for j in range (img0.shape[1]):
            img4[i][j][0]=math.floor(img3[i][j])
            img4[i][j][1]=math.floor(img3[i][j+img0.shape[1]])
            img4[i][j][2]=math.floor(img3[i][j+img0.shape[1]*2])
    end = time.perf_counter()
    #计算压缩时间和重构误差
    spent = end - start
    reconstruction_error = np.linalg.norm(img3-img1)/np.linalg.norm(img1)
    return img4,spent,compact_rate,reconstruction_error
```

```python
route = "D:/数据科学与工程算法/10205501432_温兆和_数据科学算法实验2/Images/beach/beach"
ROUTE = []
for i in range (100):
    if (i<10):
        ROUTE.append(route+str(0)+str(i)+".tif")
    else:
        ROUTE.append(route+str(i)+".tif")
```

```python
IMG_50 = []
SPENT_TIME_50 = []
COMPACT_RATE_50 = []
RECONSTRUCTION_ERROR_50 = []
for i in range(100):
    img,spent_time,compact_rate,reconstruction_error = pic_cpr(ROUTE[i],50)
    IMG_50.append(img)
    SPENT_TIME_50.append(spent_time)
    COMPACT_RATE_50.append(compact_rate)
    RECONSTRUCTION_ERROR_50.append(reconstruction_error)
```

```python
mean_spent_time_50=mean(SPENT_TIME_50)
mean_compact_rate_50=mean(COMPACT_RATE_50)
mean_reconstruction_error_50=mean(RECONSTRUCTION_ERROR_50)
sample_50 = random.randint(0,99)
print("spent_time_50")
print(mean_spent_time_50)
print("compact_rate_50")
print(mean_compact_rate_50)
print("reconstruction_error_50")
print(mean_reconstruction_error_50)
print("sample_image_50")
print(sample_50)
```

```python
Image.fromarray(np.uint8(IMG_50[sample_50]))
```

```python
img_sample_50 = tf.imread(ROUTE[sample_50])
Image.fromarray(np.uint8(img_sample_50))
```
